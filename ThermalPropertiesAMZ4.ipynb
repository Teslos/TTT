Heat capacity of the AMZ4 studied by Heinrich et al.

```python
def cpc(T):
    a = 5.225E-4
    b = 1.031E+7
    R = 8.134
    return 3*R + a * T + b * T**(-2)
def cpl(T):
    c = 0.00623
    d = -6.047E-7
    R = 8.134
    return 3*R + c * T + d * T**(-2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
temp = []
cp = []
rhoAMZ4 = 76.595
glassT = 450
for step in range(1,1200):
    if step < glassT:
        cp.append(cpl(step+273)/rhoAMZ4)
    elif step >= glassT:
        cp.append(cpc(step+273)/rhoAMZ4)
    temp.append(step)
plt.plot(temp,cp)
plt.show()
```

Heat conductivity of the AMZ4 is estimated from the values of the density, specific heat and thermal diffusivity:
$$
k = \rho C_p \alpha
$$
where $\alpha$ is the thermal diffusivity.

```python
rho=6.7
alpha=4.20E-6
def kAMZ4(T):
    if T < (glassT):
        return cpl(T+273)*rho*alpha
    elif T >= (glassT):
        return cpc(T+273)*rho*alpha
```

```python
temp = []
heat = []
for step in range(1,1200):
    if step < glassT:
        heat.append(kAMZ4(step))
    elif step >= glassT:
        heat.append(kAMZ4(step))
    temp.append(step)
plt.plot(temp,heat)
plt.show()
```
