Heat capacity of the AMZ4 studied by Heinrich et al.

```python
def cpc(T):
    a = 5.225E-4
    b = 1.031E+7
    R = 8.134
    return 3*R + a * T + b * T**(-2)
def cpl(T):
    c = 0.00623
    d = -6.047E-7
    R = 8.134
    return 3*R + c * T + d * T**(-2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
temp = []
cp = []
rhoAMZ4 = 76.595
for step in range(1,1200):
    if step < 450:
        cp.append(cpl(step+273)/rhoAMZ4)
    elif step >= 450:
        cp.append(cpc(step+273)/rhoAMZ4)
    temp.append(step)
plt.plot(temp,cp)
plt.show()
```
